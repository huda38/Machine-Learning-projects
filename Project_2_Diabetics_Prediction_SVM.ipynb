**DEPENDENCIES**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

**DATA ANALYSIS**

```python
#load the data
diabetics_data=pd.read_csv('/content/drive/MyDrive/diabetes.csv')
```

```python
#print 5rows of the data
diabetics_data.head()
```

```python
#rows&columns
diabetics_data.shape
```

```python
#statistical measuring the data
diabetics_data.describe()
```

```python
diabetics_data['Outcome'].value_counts()
```

```python
#SEPERATE DATA and label
X=diabetics_data.drop(columns='Outcome', axis=1)
Y=diabetics_data['Outcome']
print(X)
print(Y)
```

```python
scaler=StandardScaler()
scaler.fit(X)
```

```python
standardized_data=scaler.transform(X)
print(standardized_data)
```

```python
X=standardized_data
Y=diabetics_data['Outcome']

print(X)
print(Y)
```

**TRAIN & TEST**

```python
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2,stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

**TRAIN THE MODEL**

```python
classifier=svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

**MODEL EVALUATION**

```python
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy of train data: ',training_data_accuracy )
```

```python
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy of test data: ',test_data_accuracy)
```

**PREDICTIVE SYSTEM**

```python
input_data=(5,116,74,0,0,25.6,0.201,30)

#input data to numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)


#standardize input data
std_data=scaler.transform(input_data_reshaped)
print(std_data)

prediction=classifier.predict(std_data)
print(prediction)

if (prediction[0]==0):
  print('Non-diabetic patient')
else:
  print('Diabetic patient')
```
